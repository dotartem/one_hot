Импорт библиотек и модулей

````python
import pandas as pd
import random
````

Создание списка

````python
lst = ['robot'] * 10
lst += ['human'] * 10
````

Случайное перемешивание списка и создание ```DataFrame```

````python
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
````

Перевод в one hot методом ```pandas``` ```get_dummies```

````python
pd.get_dummies(data['whoAmI'])
````

Перевод в one hot вид без использования метода ```pandas``` ```get_dummies```

````python
one_hot = pd.DataFrame()
for who in data['whoAmI'].unique():
    one_hot[who] = [True if amI == who else False for amI in data['whoAmI']]
one_hot
````
